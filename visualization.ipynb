```
import  matplotlib.pyplot  as  plt
```

```
x=[3,8]  # x1 , x2 
x1=[2,7]
y=[7,14]  #  y1  , y2
y1=[8,15] 
```

```
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y,label="speed of car")  #  it will draw a line
plt.plot(x1,y1,label="speed of truck")
plt.grid(color='green')  # to draw grids 
plt.legend()
plt.show()

# this is for line form representation 
```

```
plt.xlabel("Time")
plt.ylabel("Distance")
plt.bar(x,y,label="speed of car")  #  it will draw a line
plt.bar(x1,y1,label="speed of truck")
plt.plot(x,y,label="speed of car1")  #  it will draw a line
plt.plot(x1,y1,label="speed of truck1")
plt.grid(color='green')  # to draw grids 
plt.legend()
plt.show()
```
